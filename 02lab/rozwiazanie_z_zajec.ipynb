```python
converters = {col : str.strip for col in col_names}
f = lambda x: int(str.strip(x))
num_cols = ['hours-per-week', 'age', 'capital-gain', 'capital-loss', 'fnlwgt', 'education-num']
converters.update(zip(num_cols, [f] * len(num_cols)))
converters
```

```python
df = pd.read_csv('materialy_rozpakowane/censusData/censusData', names=col_names, converters=converters)
df.replace('?', np.nan, inplace=True)
df.isna().any()
```
