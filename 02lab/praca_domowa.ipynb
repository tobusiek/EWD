```python
import pandas as pd
import numpy as np
from numpy import nan as NA

# utworzenie df
census_data = pd.read_csv('./materialy_rozpakowane/censusData/censusData', sep=', ', header=None)

# utworzenie etykiet kolumn
census_data.columns = [
    'age', 'workclass', 'fnlwgt', 'education', 'education-num', 'marital-status',
    'occupation', 'relationship', 'race', 'sex', 'capital-gain', 'capital-loss',
    'hours-per-week', 'native-country', 'salary'
]
census_data.head()
```

```python
# wyliczenie ile jest duplikatowych rekordow - malo
census_data.duplicated().value_counts()
```

```python
# pobranie i wylistowanie zduplikowanych rekordow
cd_dup = census_data[census_data.duplicated()]
cd_dup
```

```python
# wyrzucenie zduplikowanych rekordow
census_data.drop_duplicates()
```

```python
# zastapienie brakujacych wartosci wartoscia NaN
census_data = census_data.replace('?', NA)
census_data
```

```python
# sprawdzenie ile jest rekordow z przynajmniej jednym NaN
na_rows = census_data.T.isna().any()
na_rows[na_rows == True]
```

```python
# wyrzucenie rekordow z brakujacym salary
census_data = census_data.dropna(subset='salary')
census_data.head()
```

```python
# sprawdzenie ilosci rekordow z wiekiem ponizej 18
age_under18 = census_data['age'] < 18
age_under18[age_under18 == True]
```

```python
# census_data = age_under18[age_under18 == False]

# pobranie tylko tych rekordow, gdzie wiek jest przynajmniej 18
census_data = census_data.loc[census_data['age'] >= 18]
census_data.head()
```

```python
# sprawdzenie czy sa rekordy z wiekiem powyzej 120 lat
age_over120 = census_data['age'] > 120
age_over120[age_over120 == True]
```

```python
# census_data = age_over120[age_over120 == False]

# pobranie tylko tych rekordow, gdzie wiek jest ponizej 120
census_data = census_data.loc[census_data['age'] <= 120]
census_data.head()
```

```python
# po wybraniu tylko odpowiednich rekordow reset indexow
census_data.reset_index(inplace=True)
```

```python
# sprawdzenie czy sa jakies rekordy z niepoprawnym workclass
workclass = [
    'Private', 'Self-emp-not-inc', 'Self-emp-inc', 'Federal-gov', 'Local-gov',
    'State-gov', 'Without-pay', 'Never-worked'
]
validate_workclass = census_data['workclass'].isin(workclass)
validate_workclass[validate_workclass == False]
# jakies wystepuja, moze spoza listy
```

```python
# to samo dla education
education = [
    'Bachelors', 'Some-college', '11th', 'HS-grad', 'Prof-school', 'Assoc-acdm',
    'Assoc-voc', '9th', '7th-8th', '12th', 'Masters', '1st-4th', '10th', 'Doctorate',
    '5th-6th', 'Preschool'
]
validate_education = census_data['education'].isin(education)
validate_education[validate_education == False]
# wszystko dobrze
```

```python
# dla marital-status
marital_status = [
    'Married-civ-spouse', 'Divorced', 'Never-married', 'Separated', 'Widowed',
    'Married-spouse-absent', 'Married-AF-spouse'
]
validate_marital_status = census_data['marital-status'].isin(marital_status)
validate_marital_status[validate_marital_status == False]
# brak niepoprawnych rekordow
```

```python
occupation = [
    'Tech-support', 'Craft-repair', 'Other-service', 'Sales', 'Exec-managerial',
    'Prof-specialty', 'Handlers-cleaners', 'Machine-op-inspct', 'Adm-clerical',
    'Farming-fishing', 'Transport-moving', 'Priv-house-serv', 'Protective-serv',
    'Armed-Forces'
]
validate_occupation = census_data['occupation'].isin(occupation)
validate_occupation[validate_occupation == False]
# sporo bez zawodow - moze spoza listy
```

```python
relationship = [
    'Wife', 'Own-child', 'Husband', 'Not-in-family', 'Other-relative', 'Unmarried'
]
validate_relationship = census_data['relationship'].isin(relationship)
validate_relationship[validate_relationship == False]
# brak
```

```python
race = ['White', 'Asian-Pac-Islander', 'Amer-Indian-Eskimo', 'Other', 'Black']
validate_race = census_data['race'].isin(race)
validate_race[validate_race == False]
# brak
```

```python
sex = ['Female', 'Male']
validate_sex = census_data['sex'].isin(sex)
validate_sex[validate_sex == False]
# brak
```

```python
native_country = [
    'United-States', 'Cambodia', 'England', 'Puerto-Rico', 'Canada', 'Germany',
    'Outlying-US(Guam-USVI-etc)', 'India', 'Japan', 'Greece', 'South', 'China', 'Cuba',
    'Iran', 'Honduras', 'Philippines', 'Italy', 'Poland', 'Jamaica', 'Vietnam', 'Mexico',
    'Portugal', 'Ireland', 'France', 'Dominican-Republic', 'Laos', 'Ecuador', 'Taiwan',
    'Haiti', 'Columbia', 'Hungary', 'Guatemala', 'Nicaragua', 'Scotland', 'Thailand',
    'Yugoslavia', 'El-Salvador', 'Trinadad&Tobago', 'Peru', 'Hong', 'Holand-Netherlands'
]
validate_native_country = census_data['native-country'].isin(native_country)
validate_native_country[validate_native_country == False]
# wystepuja jakies spoza listy
```

```python
# wypisanie wszystkich z niepoprawnym native country, bez nana
invalid_native_country = census_data.loc[validate_native_country == False]['native-country'].notna()
idx = invalid_native_country[invalid_native_country == True].index[0]
census_data.iloc[[idx]]['native-country']
```

```python
# poprawienie wartosci 
census_data.at[idx, 'native-country'] = 'United-States'
census_data.iloc[[idx]]['native-country']
# wartosc zmieniona poprawnie
```

```python
# kategoryzacja wieku dla powyzej i ponizej 40 lat
age_category_bins = [18, 40, 120]
age_category = ['<40', '>=40']
cats = pd.cut(list(census_data['age']), age_category_bins)
cats, cats.codes
```

```python
# dodanie kolumny z kategoryzacja wieku
census_data['age-category'] = pd.Series(cats.codes)
census_data['age-category'] = census_data['age-category'].replace(range(2), list(age_category))
census_data.head()
```

```python
# przesuniecie kolumny z kat. wieku na 2 pozycje 
age_category_column = census_data.pop('age-category')
census_data.insert(1, 'age-category', age_category_column)
census_data.head()
```
